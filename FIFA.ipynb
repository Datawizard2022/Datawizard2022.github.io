```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the attribute groups
group1_attributes = ['player_id', 'player_url', 'fifa_version', 'fifa_update', 'fifa_update_date',
                     'short_name', 'long_name', 'player_positions', 'overall', 'potential',
                     'value_eur', 'wage_eur', 'age', 'dob', 'height_cm', 'weight_kg']

group2_attributes = ['club_team_id', 'club_name', 'club_position', 'club_jersey_number',
                     'club_loaned_from', 'club_joined_date', 'club_contract_valid_until_year']

group3_attributes = ['nationality_id', 'nationality_name', 'nation_team_id',
                     'nation_position', 'nation_jersey_number']

group4_attributes = ['preferred_foot', 'weak_foot', 'skill_moves', 'international_reputation',
                     'work_rate', 'body_type', 'real_face', 'release_clause_eur',
                     'player_tags', 'player_traits']
group5_attributes = [ 'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']
# Combine all the groups into a list
selected_attributes = group1_attributes + group2_attributes + group3_attributes + group4_attributes + group5_attributes

# Subset the DataFrame with selected attributes
selected_data = FIFA[selected_attributes]

# Calculate correlation matrix
correlation_matrix = selected_data.corr()
print (correlation_matrix)
# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of FIFA Player Attributes")
plt.show()
```
